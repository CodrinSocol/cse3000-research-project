```python
curve = lcdb.get_curve(16, 'sklearn.linear_model.LogisticRegression', "logloss")

def plot_curve(anchors, errors, color='green'):

    print(anchors)
    print(errors)
    n = min(len(anchors), len(errors))
    plt.scatter(anchors[0:n], errors[0:n], color=color, label='anchors')
    plt.plot(anchors[0:n], errors[0:n], color=color)
    
plt.subplots()
plot_curve(curve[0], [np.mean(x) for x in curve[1]])
plt.title('Learning Curve LogisticRegression on openml dataset 16')
plt.xlabel('Training size')
plt.ylabel('Training Error rate')
plt.legend()
plt.show()
```
